```python
import pandas as pd

fd = open('test.txt','r')
text = fd.readlines()

# colums
#Principal Diagnosis
#Secondary Diagnosis
#History of Present Illness
#Past Medical History
#Physical Examination on Discharge

'''
Discharge Date   PRINCIPAL DIAGNOSIS     SECONDARY DIAGNOSES     HISTORY OF PRESENT ILLNESS
PRE-ADMISSION MEDICATIONS    PAST MEDICAL HISTORY       FAMILY HISTORY   SOCIAL HISTORY
ALLERGIES     ADMISSION PHYSICAL EXAMINATION       STUDIES:      PROCEDURE
HOSPITAL COURSE BY PROBLEM <Subsections vary by patient>        COMPLICATIONS
CONSULTANTS         PHYSICAL EXAMINATION ON DISCHARGE
DISCHARGE MEDICATIONS.         DISPOSITION.       FOLLOW-UP APPOINTMENTS        CODE STATUS       PRIMARY CARE PHYSICIAN     
'''

pd = []
sd = []
hpi = []
pmh = []
ped = []

i = 0
while i<len(text):
    if (text[i].find('PRINCIPAL DIAGNOSIS') != -1) and (text[i].find('Processing')!= -1):
        temp = []
        temp.append(text[i])
        j = i+1
        print('a')
        while j>i and j<len(text):
            if (text[j].find('SECONDARY DIAGNOSES') != -1) and (text[j].find('Processing') != -1):
                i = j
                break
            else:
                temp.append(text[j])
                j += 1
        pd.append(temp)
            
    elif (text[i].find('SECONDARY DIAGNOSES') != -1) and (text[i].find('Processing') != -1):
        temp = []
        temp.append(text[i])
        j = i+1
        print('b')
        while j>i and j<len(text):
            if (text[j].find('HISTORY OF PRESENT ILLNESS')!=-1)  and (text[j].find('Processing') != -1):
                i = j
                break
            else:
                temp.append(text[j])
                j += 1
        sd.append(temp)        
                
    elif (text[i].find('HISTORY OF PRESENT ILLNESS') != -1) and (text[i].find('Processing') != -1):
        temp = []
        temp.append(text[i])
        print(i,'c')
        j = i+1

        while j>i  and j<len(text):
            if (text[j].find('PRE-ADMISSION MEDICATIONS') != -1 or text[j].find('PAST MEDICAL HISTORY') != -1) and (text[j].find('Processing') != -1):
                i = j
                break
            else:
                temp.append(text[j])
                j += 1
        hpi.append(temp)
            
    elif (text[i].find('PAST MEDICAL HISTORY') != -1) and (text[i].find('Processing') != -1):
        temp = []
        temp.append(text[i])
        j = i+1
        print('d')
        while j>i and j<len(text):
            if (text[j].find('FAMILY HISTORY') != -1 ) and (text[j].find('Processing') != -1):
                i = j
                break
            else:
                temp.append(text[j])
                j += 1
        pmh.append(temp)
            
    elif (text[i].find('PHYSICAL EXAMINATION ON DISCHARGE') != -1) and (text[i].find('Processing') != -1):
        temp = []
        temp.append(text[i])
        j = i+1
        print('e')
        while j>i and j<len(text):
            if text[j].find('PHYSICAL EXAMINATION ON ADMISSION') != -1 and text[j].find('Processing')  != -1:
                i = j
                break
            else:
                temp.append(text[j])
                j += 1
        ped.append(temp)  
    else:
        #print(i)
        i = i+1
print(pd[0:5])
print(pmh[0:5])
print(ped[0:5])
```
